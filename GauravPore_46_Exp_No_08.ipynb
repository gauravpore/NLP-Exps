AIM: WRITE A PROGRAM TO GENERATE PARSE TREE FROM THE TEXT AND EXTRACT NOUN AND VERB PHRASE OF ENGLISH TEXT.

THEORY: Chunking: The basic technique we will use for entity detection is chunking, which segments and labels multi-token sequences as illustrated in. The smaller boxes show the word-level tokenization and part-of-speech tagging, while the large boxes show higher-level chunking. Each of these larger boxes is called a chunk. Like tokenization, which omits whitespace, chunking usually selects a subset of the tokens. Also like tokenization, the pieces produced by a chunker do not overlap in the source text.

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
```

```python
from nltk.tokenize import PunktSentenceTokenizer

da = open("cats.txt",'r')
train_text = da.read()

sample_text = train_text
```

```python
print(train_text)
custom_sent_tokenizer = PunktSentenceTokenizer(train_text)

tokenized = custom_sent_tokenizer.tokenize(sample_text)

print(tokenized)
```

OUTPUT:

```python
def process_content():
  try:
    for i in tokenized:
      print(i)
      words = nltk.word_tokenize(i)
      tagged = nltk.pos_tag(words)
      chunkGram = r"""NLP: {<VB><.*>*?<NNP>}"""
      chunkParser = nltk.RegexpParser(chunkGram)
      chunked = chunkParser.parse(tagged)

      print(chunked)
      for subtree in chunked.subtrees(filter=lambda t: t.label() == 'Chunk'):
        print(subtree) 
        chunked.draw()
        
  except Exception as e:
     print(str(e))
process_content()
```

CONCLUSION: Hence we successfully executed the program to generate parse tree from the text and extract noun and verb phrase of English text.
