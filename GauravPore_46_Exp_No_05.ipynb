AIM: WRITE A PROGRAM TO IDENTIFY WORD FREQUENCY AND GENERATE WORD CLOUD OF ENGLISH AND HINDI TEXT

THEORY: After tokenising a text, the first figure we can calculate is the word frequency. By word frequency we indicate the number of times each token occurs in a text. When talking about word frequency, we distinguish between types and tokens. Types are the distinct words in a corpus, whereas tokens are the words, including repeats.
A word cloud is a simple yet powerful visual representation object for text processing, which shows the most frequent word with bigger and bolder letters, and with different colors. The smaller the the size of the word the lesser it’s important.
A cloud filled with lots of words in different sizes, which represent the frequency or the importance of each word. This is called Tag Cloud or WordCloud.

EXECUTION:

```python
import nltk
```

```python
from nltk.corpus import webtext
```

```python
from nltk.probability import FreqDist
```

```python
nltk.download('webtext')
```

```python
wt_words = webtext.words('/content/drive/MyDrive/Colab Notebooks/Sample_text.txt')
```

```python
data_analysis = nltk.FreqDist(wt_words)
```

```python
filter_words = dict([(m, n) for m, n in data_analysis.items() if len(m) > 3])
 
for key in sorted(filter_words):
    print("%s: %s" % (key, filter_words[key]))
 
data_analysis = nltk.FreqDist(filter_words)
 
data_analysis.plot(25, cumulative=False)
```

Program to generate Word cloud

```python
import nltk
```

```python
from nltk.corpus import webtext
```

```python
from nltk.probability import FreqDist
```

```python
from wordcloud import WordCloud
```

```python
import matplotlib.pyplot as plt
```

```python
nltk.download('webtext')
```

```python
wt_words = webtext.words('/content/drive/MyDrive/Colab Notebooks/Sample_text.txt')
```

```python
data_analysis = nltk.FreqDist(wt_words)
```

```python
filter_words = dict([(m, n) for m, n in data_analysis.items() if len(m) > 3])
wcloud = WordCloud().generate_from_frequencies(filter_words)
plt.imshow(wcloud, interpolation="bilinear")
plt.axis("off")
(-0.5, 399.5, 199.5, -0.5)
plt.show()
```
